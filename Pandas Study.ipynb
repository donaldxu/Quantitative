```python
#pd.cut(...,bins) 可以自定义自己的分类区间
mybins = [-np.inf,-7,-5,-3,0,3,5,6,np.inf]
mycats = pd.cut(nyyh.p_change,mybins)
mycats.value_counts()
```

```python
#pd.cut()经常和pd.get_dummies()函数配合使用，将连续数值离散化
p_change_dummies = pd.get_dummies(mycats,prefix='cr_dummies')
p_change_dummies.tail()
```

### 星期几是这只股票的好日子

```python
nyyh["positive"]=[ 1 if x>0 else 0 for x in nyyh.p_change]
nyyh["week"] = [x.weekday()+1 for x in nyyh.date] #0：星期一
nyyh.head(10)
```

```python
#交叉表
ct = pd.crosstab(nyyh.week,nyyh.positive)
ct
```

```python
xt_per = ct.div(ct.sum(1).astype(float),axis=0)
xt_per
```

```python
xt_per.plot(kind='bar',stacked=True)
```

```python
nyyh['preclose']=nyyh.close.shift()
nyyh.head()
```

```python
jump_df = pd.DataFrame()
jump_threshold = nyyh.close.median()*.03
def judge_jump(today):
    global jump_df
    if today.p_change>0 and (today.low-today.preclose)>jump_threshold:
        today['jump'] = 1 #向上跳空
        today['jump_power'] = (today.low-today.preclose)/jump_threshold
        jump_df = jump_df.append(today)
    elif today.p_change<0 and (today.preclose - today.high)>jump_threshold:
        today['jump'] = -1 #向下跳空
        today['jump_power'] = (today.preclose - today.high)/jump_threshold
        jump_df = jump_df.append(today)

for k_index in np.arange(0,nyyh.shape[0]):
    today = nyyh.iloc[k_index].copy()
    judge_jump(today)

jump_df
```
